```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
%matplotlib inline
```

```
train_data = pd.read_csv('https://pjreddie.com/media/files/mnist_train.csv')
```

```
train_data.head()
```

```
test_data = pd.read_csv('https://pjreddie.com/media/files/mnist_test.csv')
```

```
test_data.head()
```

```
X_train = train_data.iloc[:, 1:]
X_train.head()
```

```
y_train = train_data.iloc[:, 0]
y_train.head()
```

```
X_test = test_data.iloc[:, 1:]
X_test.head()
```

```
y_test = test_data.iloc[:, 0]
y_test.head()
```

```
nn = MLPClassifier(hidden_layer_sizes = (180,180), activation = 'logistic',solver= 'sgd')
```

```
nn.fit(X_train, y_train)
# execution took almost 13 minutes (803.8 secs)
```

```
pred = nn.predict(X_test)
print(pred)
```

```
nn.score(X_test, y_test)
```
