```python
from pydataset import data
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
%matplotlib notebook
```

```python
pima = data('Pima.tr')
```

```python
pima.head()
```

```python
pima.plot(kind = 'scatter', x = 'skin', y = 'bmi')
plt.show()
```

```python
#test train split for supervised learning (linear regression hai lite)
X_train, X_test, y_train, y_test = train_test_split(pima['skin'], pima['bmi'])
```

```python
#test train split visualisation
plt.scatter(X_train, y_train, label = 'Training Data', color = 'r'
           , alpha = 0.6)     #alpha kinda acts like opacity of the data point in graph
plt.scatter(X_test, y_test, label = 'Testing Data', color = 'b'
           , alpha = 1)
plt.legend()
plt.title('train test split')
plt.show()
```

```python
#as you can see from the plot the opacity of red and blue dots vary
```

```python
#create linear model and train it
LR = LinearRegression()  #make it a linear regression object
LR.fit(X_train.values.reshape(-1,1), y_train.values)
```

```python
#use model to predict test data
prediction = LR.predict(X_test.values.reshape(-1,1))
```

```python
#plot prediction line against actual test data
plt.plot(X_test, prediction, label = 'linear reg.', color = 'g')
plt.scatter(X_test, y_test, label = 'Actual Test Data', color = 'r', alpha = 0.7)
plt.legend()
plt.title('linear regression with test data')
plt.show()
```

```python
#predict bmi of woman with skin = 50
LR.predict(np.array([[50]]))[0]
```

```python
#score this model
LR.score(X_test.values.reshape(-1,1), y_test.values)
```
